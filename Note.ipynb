字典不是按"位置"存储，而是按**键（key）**存储。
看文章中的例子：

```python
hashmap = {1: "one", 2: "two", 3: "three"}

# 删除键 3
del hashmap[3]

# 检查删除后
if 3 in hashmap:
    print(f"Key 3 -> {hashmap[3]}")
else:
    print("Key 3 not found.")
# 输出：Key 3 not found.
```


删除键 3 后，键 3 就不存在了，不是"位置 3 空着"。你再访问 hashmap[3] 会报错，因为那个键已经没了。

和队列的区别：

数据结构	访问方式	删除后的行为
队列 queue	按顺序，队头出队	后面的元素会"顶上来"，索引变化
字典 dict	按键访问	键直接消失，其他键不受影响

```python
# 队列：删除队头后，原来的第二个元素变成新的队头
q = deque([10, 20, 30])
q.popleft()  # 删除 10
print(q[0])  # 输出：20，原来的第二个元素现在是第一个

# 字典：删除键 3 后，键 1 和 2 还是原来的键
hashmap = {1: "one", 2: "two", 3: "three"}
del hashmap[3]
print(hashmap[1])  # 输出：one，键 1 不受影响
```


所以更准确的理解是：字典是键值对映射，每个键独立存在，删除某个键不会影响其他键。

```python
# 当 n 是 10 的倍数时，计算累加和，否则返回 -1
def sum(n: int) -> int:
    if n % 10 != 0:
        return -1
    sum = 0
    # range(n + 1)：这会生成一个从 $0$ 到 $n$ 的数字序列。比如 $n=10$，它就是 $0, 1, 2, \dots, 10$。
    for i in range(n + 1): 
        sum += i
    return sum
```

```python
# 大小为 10 的数组已经装满了
arr = [i for i in range(10)]

# 现在想在数组末尾追加一个元素 10
# 需要先扩容数组
newArr = [0] * 20

# 把原来的 10 个元素复制过去
for i in range(10):
    newArr[i] = arr[i]

# 释放旧数组的内存空间
# ...

# 在新的大数组中追加新元素
newArr[10] = 10
```
